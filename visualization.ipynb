```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('../RL Challenge/highest_tile.txt', header=None)
dataset.head()
```

```python
values = dataset.values[0][1:-1]
values_int = [int(x) for x in values]
print(values_int)
```

```python
#plot the highest tile
other_vals = []
plt.plot(values_int)
plt.xlabel('episode')
plt.ylabel('highest tile')
plt.title('Highest Tile')
plt.show()
```

```python
avg_change = []
for i in range(len(values_int)):
    avg = np.mean(values_int[:i])
    avg_change.append(avg)

fig = plt.figure(figsize=(10, 6))

plt.plot(values_int)
plt.plot(avg_change)
plt.xlabel('episode')
plt.ylabel('tile value')
plt.title('Average Highest Tile')
plt.legend(['highest tile', 'average highest tile'])

f = plt.figure()
f.set_figwidth(15)
f.set_figheight(6)

plt.show()
```

```python
print(f"Mean: {np.mean(values_int)}")
print(f"Number of values: {len(values_int)}")

# print the number of unique values per value
for i in np.unique(values_int):
    print(f"{i}: {np.count_nonzero(values_int == i)}")
```
